```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1,2,3,4,5,6])
y_train=np.array([30,50,65,70,75,85])
print(f"x_train shape: {x_train.shape}")
print(f"y_train shape:{y_train.shape}")
```

```python
m=len(x_train)
print(f"No of training examples:{m}")
```

```python
plt.scatter(x_train,y_train,marker='x',c='r')
plt.xlabel("Hours studied")
plt.ylabel("Exam score")
plt.title("Exam time")
plt.show()
```

```python
slope=10
c=24
print(f"Slope={slope}")
print(f"Y-intercept={c}")
```

```python
def compute_model_output(x,slope,c):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=slope*x[i]+c

    return f_wb
```

```python
tmp_f_wb=compute_model_output(x_train,slope,c)
plt.plot(x_train,tmp_f_wb,c='b',label='Our prediction')
plt.scatter(x_train,y_train,marker='x',c='r',label='Actual Values')
plt.xlabel("Hours studied")
plt.ylabel("Exam score")
plt.title("Exam time")
plt.legend()
plt.show()
```

```python
slope=10
c=24
ExamScore=slope*7+c
print(f"Examscore={ExamScore}")
```

```python
mse=sum((100-ExamScore)**2 for i in range (len(y_train)))/len(y_train)
print(f"MSE:{mse}")
```

 Gradient Descent  Quick Notes
 What is Gradient Descent?
Gradient Descent is an optimization algorithm used to minimize a cost/loss function by updating model parameters (w, b, etc.) in the direction of the negative gradient.

# Inputs:
# x, y -> training data
# w, b -> initial parameters
# alpha -> learning rate
# num_iters -> number of iterations

for i in range(num_iters):
    dj_dw, dj_db = compute_gradient(x, y, w, b)

    # Update parameters
    w = w - alpha * dj_dw
    b = b - alpha * dj_db
def compute_gradient(x, y, w, b):
    m = len(x)
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        f_wb = w * x[i] + b
        error = f_wb - y[i]
        dj_dw += error * x[i]
        dj_db += error

    dj_dw /= m
    dj_db /= m

    return dj_dw, dj_db
